```python
import re
import time

from tqdm.notebook import tqdm

import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://guoxue123.com/jinbu/ssj/zy/'
pages = ['%03d.htm' % i for i in range(0, 11)]

data = []

for suffix in tqdm(pages):
    page = 'http://guoxue123.com/jinbu/ssj/zy/' + suffix
    
    resp = requests.get(page)
    
    resp.encoding = 'gbk'
    
    soup = BeautifulSoup(resp.text)
    

    for p in soup.select('p'):
        s = p.text.strip()

        if len(s) == 0:
            continue
            
        data.append(s)

#         if s.startswith('[疏]'):
#             data[-1]['annotations'].append(s[3:])
#         else:
#             data.append({
#                 'content': s,
#                 'annotations': []
#             })
            
    time.sleep(1)

data
```

```python
data1 = []

for line in data:
    if line.startswith('首页') or line.startswith('目录页') or \
        line.startswith('上一页') or \
        line.startswith('Powered '):
#         (len(line) <= len('上经乾传卷一') and '卷' in line):
        print(line)
    else:
        data1.append(line)
        
data1
```

```python
data2 = []

for i, s in enumerate(data1):
    if s.startswith('[疏]'):
        data2[-1]['annotations'].append(s)
    else:
        data2.append({
            'content': s,
            'annotations': []
        })

data2
```
